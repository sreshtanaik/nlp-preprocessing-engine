## Task 1: Conceptual Understanding

### 1. Difference between "Love" and "love"
In NLP, "Love" and "love" are treated as different tokens in case-sensitive systems. This increases vocabulary size unnecessarily. Converting text to lowercase ensures consistency.

### 2. What happens if stopwords are not removed?
Stopwords add noise to data and may reduce model performance. They increase processing time and do not add meaningful information.

### 3. When removing stopwords can be harmful
- In sentiment analysis: "not happy" → removing "not" changes meaning
- In question answering: removing words like "what", "where" can break meaning

### 4. Stemming vs Lemmatization
Stemming cuts words into root forms (may be incorrect), while lemmatization converts words into meaningful base forms using vocabulary.

```python
import re
from collections import Counter
```

```python
def preprocess_text(text):
    # Handle empty or invalid input
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Convert to lowercase
    text = text.lower()

    # Remove URLs and emails
    text = re.sub(r"http\S+|www\S+|@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Normalize repeated characters
    text = re.sub(r"(.)\1{2,}", r"\1\1", text)

    # Remove special characters & emojis
    text = re.sub(r"[^a-z\s]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenize
    tokens = text.split()

    # Remove short words (keep "no", "not")
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product ",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
print("===== STRESS TEST =====")

for text in test_sentences:
    tokens, cleaned = preprocess_text(text)

    print("\n--------------------")
    print("Original:", text)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
```

```python
def token_analysis(tokens):
    if len(tokens) == 0:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total

    return total, unique, round(avg_len, 2)
```

```python
print("\n===== TOKEN ANALYSIS =====")

for text in test_sentences:
    tokens, _ = preprocess_text(text)
    total, unique, avg = token_analysis(tokens)

    print("\nText:", text)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Length:", avg)
```

## Analysis

- Most noisy sentence: "Win $$$ now!!! Limited offer!!!"
- Most meaningful sentence: "I am not happy with this"

```python
all_tokens = []

for text in test_sentences:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("\nTop 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)

print("\nAll Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
print(result["clean_sentences"])
```

```python
edge_cases = ["", "😂😂😂", "123456"]

print("\n===== EDGE CASES =====")

for case in edge_cases:
    tokens, cleaned = preprocess_text(case)
    print("\nInput:", case)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
```
